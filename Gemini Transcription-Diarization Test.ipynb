```python
prompt="Transcribe this audio clip"#. Apply diarization as well"
```

```python
response = client.models.generate_content(
    #model="gemini-2.0-flash", #σε mutiple speakers σκαει
    model="gemini-2.5-flash",    # Changed from gemini-2.5-flash to gemini-1.5-pro for better stability
    contents=types.Content(
        parts=[
            types.Part(text=prompt),
            types.Part(inline_data=types.Blob(data=video_bytes, mime_type="video/mp4")),
        ]
    ),
)
```

```python
myfile = client.files.upload(file='/content/Lesson23.mp3')
prompt = 'Generate a transcript of the speech.'

response = client.models.generate_content(
  #model='gemini-2.5-flash',
  #contents=[prompt, myfile]
  model="gemini-2.5-flash",
  contents=types.Content(
    parts=[
              types.Part(text=prompt),
              types.Part(inline_data=types.Blob(data=video_bytes, mime_type="audio/mp3")),
        ]
    ),
)
```
